```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#open spin configs file
spinConfigFile = open("spinConfigs_t4.9-t4.92023_03_06_16_10_18.txt", "r")

#read in spin configs as 2D array
spinConfigs = np.loadtxt(spinConfigFile) #each row in spinConfigs is an image representing a spin configuration

#close file
spinConfigFile.close()
```

```python
#convert each line in spinConfigs to a 2D array
spinConfigs = np.array([np.reshape(spinConfigs[i], (25,25)) for i in range(len(spinConfigs))])

#plot spin configs
fig, ax = plt.subplots(1, 10, figsize=(25, 25))
for i in range(10):
    ax[i].imshow(spinConfigs[i], cmap="gray")
    ax[i].set_title("Spin Config " + str(i+1))
    ax[i].axis("off")
```
